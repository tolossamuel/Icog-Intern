the Hill Climbing algorithm problem is maximize the

f(x)=âˆ’x^2 + 4x

```python
# import library
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define the function
def f(x):
    return -x**2 + 4*x
```

```python
# random starting x value
def random_x():
    return np.random.randint(-10,10)
```

```python
def mutate_solution(current_x = 0,step_size = 0.01):
    return current_x + step_size * np.random.uniform(-1,1)
```

```python
def hill_climb(f, x_start, max_iterations=100):
    x_current = x_start
    y_current = f(x_current)
    x_path = [x_current]
    y_path = [y_current]

    for _ in range(max_iterations):
        x_new = x_current + mutate_solution(x_current)
        x_new = np.clip(x_new, 0, 4)
        y_new = f(x_new)

        if y_new > y_current:
            x_current = x_new
            y_current = y_new
            x_path.append(x_current)
            y_path.append(y_current)

    return x_current, y_current, x_path, y_path
```

```python
x_start = random_x()
x_opt, y_opt, x_path, y_path = hill_climb(f, x_start)
```

```python
x_vals = np.linspace(0, 4, 400)
y_vals = f(x_vals)
```

```python
plt.plot(x_vals, y_vals, label="Function f(x)", color="blue")
plt.plot(x_path, y_path, 'ro-', label="Hill Climbing Path")
plt.scatter(x_opt, y_opt, color='green', s=100, label=f"Optimum: x={x_opt:.2f}")
plt.title("Hill Climbing Optimization")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```
